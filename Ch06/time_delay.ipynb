```python
# prompt: generate cos signal with downsampling by 2

import numpy as np
import matplotlib.pyplot as plt

# Original signal parameters
sampling_rate = 100  # Samples per second
duration = 1  # Seconds
frequency = 5  # Hz

# Generate the original cosine signal
t = np.arange(0, duration, 1/sampling_rate)
original_signal = np.cos(2 * np.pi * frequency * t)

# Downsample the signal by a factor of 2
downsampled_signal = original_signal[::2]
downsampled_time = t[::2]


# Plot the signals
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, original_signal)
plt.title('Original Cosine Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(downsampled_time, downsampled_signal)
plt.title('Downsampled Cosine Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')


plt.tight_layout()
plt.show()
```

```python
# prompt: generate cos with frequency=2, range=[0,2], phase angle = pi/4, sampling frequency=500Hz

import numpy as np
import matplotlib.pyplot as plt

# Signal parameters
frequency = 2  # Hz
duration = 2  # seconds
phase_angle = np.pi / 4  # radians
sampling_frequency = 500  # Hz

# Time vector
t = np.arange(0, duration, 1/sampling_frequency)
delay_sample = 100 # Hz
delay_time = delay_sample/sampling_frequency
time_delay = t + delay_time

# Generate the cosine signal
signal = np.cos(2 * np.pi * frequency * t + phase_angle)
#delay_signal = np.cos(2 * np.pi * frequency * time_delay + phase_angle)

# Plot the signal
plt.figure(1)
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Cosine Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(2, 1, 2)
plt.plot(time_delay, signal)
#plt.xlim(0+delay_time, 2)
plt.title(f'Cosine Signal with time delay = {delay_time}sec')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()

plt.figure(2)
plt.plot(t, signal, 'b--')
plt.plot(time_delay, signal, 'r-')
plt.legend(['Original Signal', 'Delayed Signal'])
plt.title(f'Cosine Signal with time delay = {delay_time}sec')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# prompt: plot cos with delay=0.2

import numpy as np
import matplotlib.pyplot as plt

# Signal parameters
frequency = 2  # Hz
duration = 2  # seconds
phase_angle = np.pi / 4  # radians
sampling_frequency = 500  # Hz
delay_time = 0.2

# Time vector
t = np.arange(0, duration, 1/sampling_frequency)

# Generate the cosine signal
signal = np.cos(2 * np.pi * frequency * t + phase_angle)

# Time vector for the delayed signal
delayed_t = t + delay_time

# Generate the delayed cosine signal
delayed_signal = np.cos(2 * np.pi * frequency * (t + delay_time) + phase_angle)


# Plot the signals
plt.figure(figsize=(10, 6))
plt.plot(t, signal, label='Original Signal')
plt.plot(delayed_t, delayed_signal, label=f'Delayed Signal (delay = {delay_time} s)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Cosine Signal with Delay')
plt.legend()
plt.grid(True)
plt.show()
```
