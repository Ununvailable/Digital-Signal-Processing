```python
# prompt: plot cos downsampling by 2

import numpy as np
import matplotlib.pyplot as plt

# Original signal
fs = 100  # Sampling frequency
t = np.arange(0, 1, 1/fs)
x = np.cos(2 * np.pi * 5 * t)

# Downsampling by a factor of 2
downsampled_x = x[::2]
downsampled_t = t[::2]


# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='Original Signal')
plt.plot(downsampled_t, downsampled_x, 'ro-', label='Downsampled Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave Downsampling')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: plot sin downsampling by 2 with average value of each sample

import numpy as np
import matplotlib.pyplot as plt

# Original signal
fs = 100  # Sampling frequency
t = np.arange(0, 1, 1/fs)
x = np.cos(2 * np.pi * 5 * t)

# Downsampling by a factor of 2 with averaging
downsampled_x = []
downsampled_t = []
for i in range(0, len(x), 2):
    if i + 1 < len(x):
        downsampled_x.append(np.mean(x[i:i+2]))
        downsampled_t.append(np.mean(t[i:i+2]))
    else:
        downsampled_x.append(x[i]) # handle the last sample if the length is odd
        downsampled_t.append(t[i])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t, x, label='Original Signal')
plt.plot(downsampled_t, downsampled_x, 'ro-', label='Downsampled Signal (Averaged)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Cosine Wave Downsampling with Averaging')
plt.legend()
plt.grid(True)
plt.show()
```
